```python
pokemon.Name = pokemon.Name.astype('category')
pokemon.Type_1 = pokemon.Type_1.astype('category')
pokemon.Type_2 = pokemon.Type_2.astype('category')
pokemon.Color = pokemon.Color.astype('category')
pokemon.Egg_Group_1 = pokemon.Egg_Group_1.astype('category')
pokemon.Egg_Group_2 = pokemon.Egg_Group_2.astype('category')
pokemon.Body_Style = pokemon.Body_Style.astype('category')
```

```python
k1 = sns.lmplot(data = pokemon, x = 'HP', y = 'Attack', hue = 'Egg_Group_1', fit_reg = False )
k2 = sns.lmplot(data = pokemon ,x = 'HP', y = 'Attack', hue = 'Egg_Group_2', fit_reg = False)
```

```python
f, axes = plt.subplots(1,2, figsize = (12,6))
v1 = sns.violinplot(data = pokemon, y = 'Type_1' , x = 'Attack', width = 1,gridsize = 200 , ax = axes[0] )
v2 = sns.violinplot(data = pokemon, y = 'Type_2' , x = 'Attack', width = 1,gridsize = 100, ax = axes[1])
```
